```python
import folium
import pandas
```

```python
df1 = pandas.read_excel("/Users/jonasmiosga/Documents/UniversitaÌˆt_Konstanz/project/datasets_coord/Asia_2010-2019_Sep28.xlsx")
```

```python
# extract drug-related events, year 2018
df_drug = df1[df1.NOTES.str.contains("drug")]
df_drug = df_drug[df_drug.YEAR == 2018]
```

```python
# create list of coordinates for folium
coord_notes_lst = []
for i in range(0, len(df_drug)):
    coord_notes_lst.append([float(df_drug["LATITUDE"][i:i+1]), float(df_drug["LONGITUDE"][i:i+1]), df_drug["NOTES"][i:i+1].iloc[0]])
```

```python
map = folium.Map(location = [21.348317, 88.602014], zoom_start = 4)
fg = folium.FeatureGroup(name = "My Map")

for i in range(len(coord_notes_lst)):
    fg.add_child(folium.Marker(location = coord_notes_lst[i][0:2], radius = 2, popup = coord_notes_lst[i][2]))

map.add_child(fg)
map.save("map.html")
map
```
